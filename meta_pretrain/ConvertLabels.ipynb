```python
from os import walk
from tqdm import tqdm
from PIL import Image
import torch
from tqdm import tqdm
import glob
import numpy as np
import json
import numpy as np
```

```python
with open('../construction_images/data/test.json') as json_data:
    labels = json.load(json_data)
    json_data.close()
```

```python
categories = ["Deckenansicht", "Detail", "Revisionsoeffnung", "Wandansicht", "Kabel", "Baugeraet", "Trockenbau", "Tuer", "Fertigstellung", "Brandschutz", "Steckdose_Schalter", "Beleuchtung", "Lueftung", "Fenster", "Baumaterial", "Aussenbereich", "Trasse", "Buero", "Rohbau", "Betonwand", "Bauschutt", "Abwasser", "Heizungsrohr", "Durchbruch", "Erschliessung", "Brandmelder", "Keller", "Mauerwerk", "Sprinkler", "Baunotiz", "Installation", "Bad", "Deckenheizung", "Technikzentrale", "Parkhaus", "Schacht", "Bodenansicht"]
```

```python
for image in labels.keys():
    image_categories = []
    for i, v in enumerate(labels[image]):
        if v == 1:
            image_categories.append(categories[i])
    labels[image] = {}
    labels[image]['labels'] = image_categories 
```

```python
tasks = []

for category in categories: 
    task = {}
    task['label'] = category
    task['images-positive'] = []    
    task['images-negative'] = []
    
    for image in labels.keys():
        if category in labels[image]['labels']:
            task['images-positive'].append(image)
        else:
            if len(task['images-negative']) < 30:
                task['images-negative'].append(image)
                
    tasks.append(task)
```

```python
with open('tasks_test.json', 'w') as fp:
    json.dump(tasks, fp)
tasks
```
